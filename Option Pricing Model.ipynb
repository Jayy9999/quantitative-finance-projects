```python
# Call option pricing using black scholes model

import numpy as np
from scipy import stats
import datetime

# S = stock price, K = Strike price, T = Time to maturity, r = RFR, v = Vol

def bscalloption(S,K,T,r,v):
    d1 = np.log(S/K) + (r+v*v/2)*T/v*np.sqrt(T)
    d2 = d1 - v*np.sqrt(T)
    return S*stats.norm.cdf(d1) - K*np.exp(-r*T)*stats.norm.cdf(d2)

S = 160.65
K = 170.00
T = 34/365
r = 0.05
v = 0.3541

price = bscalloption(S,K,T,r,v)
print(price)
```
